```python
from numpy import pi, arange, cos, sum, cumsum, diff, roll


def enum_print(arr):
    for index, value in enumerate(arr):
        print(f"{index:2}  {value:.4g}")
    return
```

```python
# Exemplo capítulo 22
# D = 2 * 381e-3
# Mx = 200e3
# F = 100e3
# d = 150e-3
# z = arange(16)
# t_skin = 0.8e-3
# w = 1e-3
# h = 100e-3
# t_stringer = 1e-3

# Valores da Opção
D = 880e-3
Mx = 235e3
F = 120e3
d = 170e-3
z = arange(16)
t_skin = 0.8e-3
w = 26e-3
h = 30e-3
t_stringer = 2e-3
```

```python
y = D / 2 * cos(z * pi / 8)
enum_print(y * 1e3)
```

```python
A_stringer = (w + h - t_stringer) * t_stringer
print(A_stringer * 1e6)

B = A_stringer + t_skin * D * pi / 16 / 6 * (4 + (roll(y, 1) + roll(y, -1)) / y)
# B = B*0 + B[1]
enum_print(B * 1e6)
```

```python
Ixx = sum(B * y ** 2)
print(f"{Ixx * 1e12:.4g}")
```

```python
sigma_z = Mx * y / Ixx
enum_print(sigma_z * 1e-6)
```

```python
qf = F / Ixx * cumsum(B * y)
print((qf[0] + qf[-1])/2)
qf = qf - (qf[0] + qf[-1]) / 2
enum_print(qf * 1e-3)
```

```python
qt = F * d / (pi * D * D / 2)
print(qt * 1e-3)
```

```python
q = qf + qt
enum_print(q * 1e-3)
```

```python
print(sum(diff([*y, y[0]]) * q))
```
