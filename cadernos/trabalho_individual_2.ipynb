Primeiro vamos criar todos os símbolos fornecidos pela questão e configurar as respectivas substituições numéricas paraencontrar valores no final

```python
# Constantes comuns
t = sp.symbols("t", real=True)
num_subs = {t: 2}

# Constantes da opção
X1, Y1, a, b, c = sp.symbols("X_1, Y_11, a, b, c", real=True)
num_subs.update({X1: 105, Y1: 205, a: 22, b: 105, c: 82})  # Opção I
```

Precisamos também criar os símbolos intermediários que desejamos encontrar

```python
# Variáveis e incógnitas
center = sp.Matrix(sp.symbols("x_c, y_c", real=True))
M = sp.Matrix(sp.symbols("M_x, M_y"), real=True)
I = sp.Matrix(2, 2, sp.symbols("I_xy, I_xx, I_yy, I_xy"), real=True)
```

Então, configuramos novos valores para as variáveis intermediárias. Para preservar os valores originais e permitir inspeção mais fácil das expressões, criaremos essas relações como um dicionário de substituições numéricas

O primeiro valor que precisamos é a posição do centróide da seção. Aqui, calculamos sua posição a partir de um ponto arbitrário no canto inferior esquerdo da seção.

```python
# Sub áreas da seção
A = a * t, (b - 2 * t) * t, c * t

# coordenadas do centro
sym_subs = {
    center[0]: sp.simplify((A[0] * a / 2 + A[1] * t / 2 + A[2] * c / 2) / sum(A)),
    center[1]: sp.simplify((A[0] * (b - t / 2) + A[1] * b / 2 + A[2] * t / 2) / sum(A)),
}

# Exibe as substituições realizadas e valores numéricos
math(f"(\\bar{{x}}, \\bar{{y}}) = {sp.latex(center)}")
math(f"(\\bar{{x}}, \\bar{{y}}) = " + sp.latex(center.subs(sym_subs)))
```

Então, calculamos os momentos de Inércia:

```python
sym_subs.update(
    {
        "I_xx": (
            A[0] * t ** 2 / 12
            + A[0] * (b - t / 2 - center[1]) ** 2
            + A[1] * (b - 2 * t) ** 2 / 12
            + A[1] * (b / 2 - center[1]) ** 2
            + A[2] * t ** 2 / 12
            + A[2] * (t / 2 - center[1]) ** 2
        ),
        "I_yy": (
            A[0] * a ** 2 / 12
            + A[0] * (a / 2 - center[0]) ** 2
            + A[1] * t ** 2 / 12
            + A[1] * (t / 2 - center[1]) ** 2
            + A[2] * c ** 2 / 12
            + A[2] * (c / 2 - center[1]) ** 2
        ),
        "I_xy": (
            A[0] * (a / 2 - center[0]) * (b - t / 2 - center[1])
            + A[1] * t * (t / 2 - center[0]) * (b / 2 - center[1])
            + A[2] * (c / 2 - center[0]) * (t / 2 - center[1])
        ),
    }
)

math(f"{sp.latex(I)} = {sp.latex(sp.simplify(I.subs(sym_subs)))}")
math(f"{sp.latex(I)} = {sp.latex(I.subs(sym_subs).evalf(5, subs=num_subs))}")
```

```python
alpha, E, rho, z = sp.symbols("alpha, E, rho, z", real=True)
num_subs[z] = 2000
sym_subs.update({
    "M_x": -Y1 * z,
    "M_y": X1 * z
})
math(sp.latex(M) + "=" + sp.latex(M.subs(sym_subs)))
math(sp.latex(M) + "=" + sp.latex(M.subs(sym_subs).evalf(5, subs=num_subs)))
```

```python
rot = (I.inv() * M)
rot.subs(sym_subs).evalf(5, num_subs)
```

```python
tan_alpha = sp.simplify(rot[0]/rot[1])
math(r"\tan\alpha = " + sp.latex(tan_alpha))
math(r"\tan\alpha = " + sp.latex(tan_alpha.subs(sym_subs).evalf(5, subs=num_subs)))
```

```python
(sp.atan2(rot[0], rot[1])* 180 / sp.pi).subs(sym_subs).evalf(5, subs=num_subs)
```

```python
xp, yp = sp.symbols("x_p, y_p")
p = sp.Matrix([xp - center[0], yp - center[1]])
sigma = p.T * rot

p_of_interest = (
    (0, 0),
    (0, b),
    (a, b),
    (a, b - t),
    (c, 0),
    (c, t),
)

for coords in p_of_interest:
    s_ = (
        sigma.subs({xp: coords[0], yp: coords[1]})
        .subs(sym_subs)
        .evalf(5, subs=num_subs)
    )
    math(f"\\sigma_{{{sp.latex(coords)}}} = " + sp.latex(s_))

```
