```python
# Problem properties (option V)
a = 42e-3
b = 82e-3
ta = 1.2e-3
tb = 2.2e-3
Sx = 1e3
```

```python
# Moments of inertia
Ixx = 2 * b ** 3 * tb / 12 + 2 * a * ta * (b / 2) ** 2
Iyy = 2 * a ** 3 * ta / 12 + 2 * b * tb * (a / 2) ** 2

print(Ixx * 1e12, Iyy * 1e12) # [mm^4]
```

```python
# Shear flow
qk = -Sx / Iyy
s1 = 0
q1 = 0
s2 = b / 2
q2 = q1 + qk * tb * (-a / 2) * s2
s3 = b / 2
q3 = q2 + qk * ta * s3 * (s3 / 2 - a / 2)

print(q2 / 1e3, q3 / 1e3) # [N/m]
```
