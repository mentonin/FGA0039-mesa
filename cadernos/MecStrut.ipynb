## Stress Equations

### Equilibrium Equations

$$
\frac{∂ σ_x}{∂ x} + \frac{∂ τ_{xy}}{∂ y} + \frac{∂ τ_{xz}}{∂ z} + X = 0
$$
$$
\frac{∂ σ_y}{∂ y} + \frac{∂ τ_{yx}}{∂ x} + \frac{∂ τ_{yz}}{∂ z} + Y = 0
$$
$$
\frac{∂ σ_z}{∂ z} + \frac{∂ τ_{zx}}{∂ x} + \frac{∂ τ_{zy}}{∂ y} + Z = 0
$$

### Boundary Conditions

$$
\overline{X} = σ_x l + τ_{xy} m + τ_{xz} n
$$
$$
\overline{Y} = σ_y m + τ_{yz} n + τ_{yx} l
$$
$$
\overline{Z} = σ_z n + τ_{zx} l + τ_{zy} m
$$
where $l$, $m$, $n$ are the direction cosines between the normal to the surface and axis $x$, $y$, $z$, respectively

## Strain-Deformation Equations

### Longitudinal/Direct Strain

$$
ϵ = \lim\limits_{L\to 0} \frac{ΔL}{L}
$$
$$
ϵ_x = \frac{∂ u}{∂ x}
$$
$$
ϵ_y = \frac{∂ v}{∂ y}
$$
$$
ϵ_z = \frac{∂ w}{∂ z}
$$


### Shear Strain

$$
γ_{xy} = \frac{∂ v}{∂ x} + \frac{∂ u}{∂ y}
$$
$$
γ_{xz} = \frac{∂ w}{∂ x} + \frac{∂ u}{∂ z}
$$
$$
γ_{yz} = \frac{∂ w}{∂ y} + \frac{∂ v}{∂ z}
$$

### Compatibility Equations

$$
\frac{∂^2 γ_{xy}}{∂x ∂y} = \frac{∂^2 ϵ_y}{∂x^2} + \frac{∂^2 ϵ_x}{∂y^2}
$$
$$
\frac{∂^2 γ_{yz}}{∂y ∂z} = \frac{∂^2 ϵ_z}{∂y^2} + \frac{∂^2 ϵ_y}{∂z^2}
$$
$$
\frac{∂^2 γ_{xz}}{∂x ∂y} = \frac{∂^2 ϵ_z}{∂x^2} + \frac{∂^2 ϵ_x}{∂z^2}
$$

Or, rearranged,
$$
2 \frac{∂^2 ϵ_x}{∂y ∂z} = \frac{∂}{∂x} \Bigg( - \frac{∂γ_{yz}}{∂x} + \frac{∂γ_{xz}}{∂y} + \frac{∂γ_{xy}}{∂z} \Bigg)
$$
$$
2 \frac{∂^2 ϵ_y}{∂x ∂z} = \frac{∂}{∂y} \left( \frac{∂γ_{yz}}{∂x} - \frac{∂γ_{xz}}{∂y} + \frac{∂γ_{xy}}{∂z} \right)
$$
$$
2 \frac{∂^2 ϵ_z}{∂x ∂y} = \frac{∂}{∂z} \left( \frac{∂γ_{yz}}{∂x} + \frac{∂γ_{xz}}{∂y} - \frac{∂γ_{xy}}{∂z} \right)
$$

## Stress-Strain Relationship

$$
ϵ_x = \frac{1}{E}[ σ_x - ν( σ_y + σ_z ) ]
$$
$$
ϵ_y = \frac{1}{E}[ σ_y - ν( σ_x + σ_z ) ]
$$
$$
ϵ_z = \frac{1}{E}[ σ_z - ν( σ_x + σ_y ) ]
$$

$$
σ_x = \frac{νE}{(1+ν)(1-2ν)}e + \frac{E}{(1+ν)}ϵ_x
$$
$$
σ_y = \frac{νE}{(1+ν)(1-2ν)}e + \frac{E}{(1+ν)}ϵ_y
$$
$$
σ_z = \frac{νE}{(1+ν)(1-2ν)}e + \frac{E}{(1+ν)}ϵ_z
$$
$$
e = ϵ_x + ϵ_y + ϵ_z
$$

$$
γ_{xy} = \frac{τ_{xy}}{G}
$$
$$
γ_{yz} = \frac{τ_{yz}}{G}
$$
$$
γ_{xz} = \frac{τ_{xz}}{G}
$$
$$
G = \frac{E}{2(1+ν)}

## Plane Stress

For thin sheet components, the stresses across the thickness of the sheet are usually negligible, so we can simplify the equation system using the following assumptions:
$$
σ_z = τ_{zx} = τ_{zy} = 0
$$

### Plane Stress Equilibrium Equations

$$
\frac{∂ σ_x}{∂ x} + \frac{∂ τ_{xy}}{∂ y} + X = 0
$$
$$
\frac{∂ σ_y}{∂ y} + \frac{∂ τ_{yx}}{∂ x} + Y = 0
$$

### Plane Stress Stress-Strain Relationship

$$
ϵ_x = \frac{1}{E}( σ_x - ν σ_y )
$$
$$
ϵ_y = \frac{1}{E}( σ_y - ν σ_x )
$$
$$
ϵ_z = \frac{-ν}{E}( σ_x + σ_y )
$$

$$
σ_x = \frac{E}{1-ν^2} (ϵ_x + ν ϵ_y)
$$
$$
σ_y = \frac{E}{1-ν^2} (ϵ_y + ν ϵ_x)
$$

## Plane Strain

Similarly, if displacements are assumed to happen only in one plane, we have the plane strain condition.
$$
ϵ_z = γ_{zx} = γ_{zy} = 0
$$

In the plane strain condition, we can simplify the compatibility equations to:
$$
\frac{∂^2 γ_{xy}}{∂x ∂y} = \frac{∂^2 ϵ_{y}}{∂x^2} + \frac{∂^2 ϵ_{x}}{∂y^2}
$$
