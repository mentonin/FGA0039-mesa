```python
import sympy as sp
```

```python
L, a, b, h, t1, t2, t3 = sp.symbols("L, a, b, h, t_1, t_2, t_3")
xc, yc = sp.symbols("x_c, y_c")
E = sp.symbols("E")

subs_num = {
    L: 450e-3,
    a: 10e-3,
    b: 10e-3,
    h: 18e-3,
    t1: 4e-3,
    t2: 3e-3,
    t3: 5e-3,
    E: 2e11,
}
c = h - t1 - t3

milli = 1000
```

```python
A = a * t1 + b * t3 + c * t2
print("Area:", A.evalf(5, subs=subs_num) * milli ** 2)

xc = (a * t1 * (b - t2 + a / 2) + b * t3 * (b / 2) + t2 * c * (b - t2 / 2)) / A
yc = (a * t1 * (h - t1 / 2) + b * t3 * (t3 / 2) + t2 * c * (t3 + c / 2)) / A
print("xc:", xc.evalf(5, subs=subs_num) * milli)
print("yc:", yc.evalf(5, subs=subs_num) * milli)
```

```python
Ixx = (
    a * t1 ** 3 / 12
    + a * t1 * (h - t1 / 2 - yc) ** 2
    + b * t3 ** 3 / 12
    + b * t3 * (t3 / 2 - yc) ** 2
    + t2 * c ** 3 / 12
    + t2 * c * (t3 + c / 2 - yc) ** 2
)

Iyy = (
    a ** 3 * t1 / 12
    + a * t1 * (b - t2 + a / 2 - xc) ** 2
    + b ** 3 * t3 / 12
    + b * t3 * (b / 2 - xc) ** 2
    + t2 ** 3 * c / 12
    + t2 * c * (b - t2 / 2 - xc) ** 2
)

Ixy = (
    +a * t1 * (b - t2 + a / 2 - xc) * (h - t1 / 2 - yc)
    + b * t3 * (b / 2 - xc) * (t3 / 2 - yc)
    + t2 * c * (b - t2 / 2 - xc) * (t3 + c / 2 - yc)
)

print("Ixx:", Ixx.evalf(5, subs=subs_num) * milli ** 4)
print("Iyy:", Iyy.evalf(5, subs=subs_num) * milli ** 4)
print("Ixy:", Ixy.evalf(5, subs=subs_num) * milli ** 4)
```

```python
I_min = (Ixx + Iyy) / 2 - sp.sqrt((Ixx - Iyy) ** 2 / 4 + Ixy ** 2)
print("I_min:", I_min.evalf(5, subs=subs_num) * milli ** 4)
```

```python
Pcr = sp.pi ** 2 * E * I_min / (2 * L) ** 2
print("Pcr:", Pcr.evalf(5, subs=subs_num))
```

```python
theta = sp.atan(2 * Ixy / (Iyy - Ixx)) / 2 * 180 / sp.pi
print("theta:", theta.evalf(5, subs=subs_num))
```
